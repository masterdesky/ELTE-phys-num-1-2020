A hangot Pandorka okostelefonjával rögzítette és egy `signal.npy` nevű `numpy` bináris file formájában

A file egy két oszlopos `array`-t tartalmaz. Az első oszlop a hang kezdetétől eltet idő a második oszlop pedig maga a hang. 

- Olvassuk be az adatfile-t és kézítsünk egy ábrát a hang időbeni lefutásáról!
- Készítsünk egy ábrát a jel Fourier-spektrumáról is, pozitív frekvenciákra! Az ábrán jelöljük azt a frevenciát amely a spektrum maximumához tartozik.
- Határozzuk meg a rádiójel spektrogrammját is! 
- A spektogramm alapján diszkutáld saját szavaiddal, hogy miután a hang elcsendesedett szerinted melyik testvérét látta viszont Pandorka!

```python
import numpy
from pylab import *
from scipy.fftpack import fft,fftfreq
import matplotlib.pyplot as plt
from scipy.signal import spectrogram #feladat megoldásához szükséges importok
```

```python
signal_data = load('signal.npy') #betöltjük a filet
```

```python
signal_data #valóban egy kétoszlopos array
```

```python
t = signal_data[:, 0]     #az eltelt idő a signal_data első oszlopa
voice = signal_data[:, 1] #maga a hang a második oszlop
```

```python
plot(t, voice, color ='hotpink') #ábrázoljuk a hangot az idő függvényében, rózsaszínben
```

```python
Fsignal = fft(voice)                         #legyártjuk fast Fourier transzformációját a jelnek
freq = fftfreq(len(voice),voice[1]-voice[0]) #legyártjuk a frekvenciát a hang teljes hosszára 

plot(freq,abs(Fsignal), color = 'c')         #plottoljuk a transzformációt a frekvencia függvényében
```

```python
fs = 1/t[-1]                             #mintavételezési frekvencia, ilyen távol lesznek a pontok egymástól
freqs, t, retek = spectrogram(voice ,fs) #elkészítjük magát a spektrogrammot
pcolor(t,freqs,retek)                    #plottoljuk 
```

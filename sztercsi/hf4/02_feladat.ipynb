```python
import pandas as pd
%pylab inline
from difflib import SequenceMatcher
```

```python
df = pd.read_excel('metadata.xlsx') #betölrjük a filet
new_df = df.dropna(subset=['Age', 'Diagnosis', 'Gender', 'Date of sample collection']) #töröljük azokat a sorokat, amik a feladat szerint nem kellenek
new_df #így ezt kapjuk
```

A szelektálásra ugyanazt a módszert fogom használni mind a Gender mind a Diagnosis oszlopra
- csinálok egy listát, amiben a helyesen leírt szavak vannak(amiket keresünk)
- csinálok egy egy listát a vizsgált oszlopból
- továbbá egy üres dictet

Ezek után a SequenceMatcher függvénnyel készítek egy új függvényt, majd erre írok egy for ciklust, ami végig iterál a helyes szavak listáján és összehasonlítja egyesével az oszlopból készített lista elemeivel. Ennek az eredménye mindig egy 0 és 1 közti szám lesz, ami megmutatja, hogy mennyire hasonlít a két szó. A dictet úgy csinálom, hogy a vizsgált szó- egyesével a dataframe összes szavához a hasonlósági érték.

Majd megnézem, hogy kb milyen érték felett beszélünk javítandó szóról, és csak ezeket javítom, majd ebből készített listát alakítom vissza a dataframe egyik oszlopává.

```python
good_words = ['Prostate adenocarcinoma', 'Lung Squamous Cell Carcinoma', 'Lung adenocarcinoma', 'Colon adenocarcinoma', 'Breast Invasive Carcinoma']
DfToList = list(new_df['Diagnosis'])

sim= {}

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

for good in good_words:
    sim_list = []
    for word in DfToList:
        comp = similar(good, word)
        sim_list.append(comp)
    sim[good] = sim_list
```
